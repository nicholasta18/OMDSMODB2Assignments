# 4. Storytelling With Data plot

Reproduce any graph of your choice in p. 187-205 of the Storytelling With Data book as best you can. (Chapter Eight). You do not have to get the exact data values right, just the overall look and feel.

```python
# Create figure and axes
fig, ax = plt.subplots(figsize=(8, 6))

# Data Generation
years = np.arange(2008, 2015)

product_a_prices = np.array([395, 415, 420, 400, 315, 270, 260])
product_b_prices = np.array([360, 400, 410, 390, 290, 255, 245])
product_c_prices = np.array([np.nan, np.nan, 100, 180, 200, 220, 200])
product_d_prices = np.array([np.nan, np.nan, np.nan, 170, 260, 210, 250])
product_e_prices = np.array([np.nan, np.nan, np.nan, np.nan, np.nan, 100, 200])

# Plotting the lines
products = {
    'Product A': {'data': product_a_prices, 'color': '#336699'},
    'Product B': {'data': product_b_prices, 'color': '#cc3333'},
    'Product C': {'data': product_c_prices, 'color': '#669933'},
    'Product D': {'data': product_d_prices, 'color': '#996699'},
    'Product E': {'data': product_e_prices, 'color': '#339999'}
}

for product, info in products.items():
    ax.plot(years, info['data'], label=product, color=info['color'], linewidth=2)

ax.set_title('Average Retail Product Price per Year', fontsize=16)
fig.text(0.9, 0.94, 'Lesson 3: eliminate clutter', ha='right', va='top', fontsize=12, color='gray')
ax.set_xticks(years)
ax.set_xlabel('')
ax.set_ylabel('Price', fontsize=12)
ax.set_ylim(-10, 500)
y_ticks = np.arange(0, 501, 100)
ax.set_yticks(y_ticks)
ax.set_yticklabels([f'${tick}' for tick in y_ticks])
ax.grid(True, linestyle='-', which='major', color='gray', alpha=0.5)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), fancybox=True, shadow=False, ncol=5)
fig.text(0.5, 0.0, 'FIGURE 8.5 Single line graph for all products', ha='center', va='bottom', fontsize=12, color='black')
plt.tight_layout(rect=[0, 0.05, 1, 0.95]) 
plt.show()


```
